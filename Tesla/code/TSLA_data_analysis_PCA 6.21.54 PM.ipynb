```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
concatenated_data = pd.read_excel("concatenated_table_1116_logistic.xlsx",index_col=0) 
#same data as used in logistic model will be used here
```

```python
concatenated_data.head()
```

```python
concatenated_data
```

```python
concatenated_data.dropna(axis=1,inplace=True)
concatenated_data
```

```python
concatenated_data.drop(['Change_stock_prices'],axis=0,inplace =True)
```

```python
concatenated_data
```

```python
# Removing words that don't appear for atleast 50 days or are smaller than two characters
minimum_occurance = 100 #minimum number of days a word should appear
minimum_word_length = 3
rare_short_words = []
for idx in concatenated_data.index[:-1]:
    if ((concatenated_data.loc[idx] == 0).sum()<minimum_occurance) or (len(idx)<minimum_word_length):
        rare_short_words.append(idx)
```

```python
TSLA_data_pca = []
TSLA_data_pca = concatenated_data.drop(rare_short_words)
```

```python
TSLA_data_pca = TSLA_data_pca.transpose()
TSLA_data_pca
```

## Principal Component Analysis :

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(TSLA_data_pca)
```

```python
x_pca = pca.transform(TSLA_data_pca)
```

```python
TSLA_data_pca.shape
```

```python
x_pca.shape
```

## Visualization <-- Will change with n_components selected

```python
concatenated_data_pca = concatenated_data.transpose()
concatenated_data_pca['Change_binary']
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')

ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=concatenated_data_pca['Change_binary'],cmap='plasma')
ax.set_xlabel('First principal component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
```

### First, second, third PCs are weighted sums of feature, they don't represent which 3 features have most weight. We can check that by plotting a heat map though. 

### Heat map to check relative weights of each feature in a PC. Though, it's impractical for 1000s for features. (Plot will look messed) 

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=TSLA_data_pca.columns)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```
