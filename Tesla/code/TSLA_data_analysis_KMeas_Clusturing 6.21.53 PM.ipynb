### K means is unsupervised learning method. We are using this just to check if our data has any clusters to start with. In reality, we will never use unsupervised method as we do. have labels.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
concatenated_data = pd.read_excel("concatenated_table_1116_logistic.xlsx",index_col=0) 
#same data as used in logistic model will be used here
```

```python
concatenated_data.head()
```

```python
concatenated_data
```

```python
concatenated_data.dropna(axis=1,inplace=True)
concatenated_data
```

```python
#removing conrrect labels "Change_binary" to use concatenated_data_kmc in KMeans which is unsupervised learning. 
concatenated_data_kmc = concatenated_data.drop(['Change_stock_prices','Change_binary'],axis=0)
```

```python
concatenated_data_kmc
```

```python
# Removing words that don't appear for atleast 50 days or are smaller than two characters
minimum_occurance = 50 #minimum number of days a word should appear
minimum_word_length = 3
rare_short_words = []
for idx in concatenated_data_kmc.index:
    if ((concatenated_data_kmc.loc[idx] == 0).sum()<minimum_occurance) or (len(idx)<minimum_word_length):
        rare_short_words.append(idx)
```

```python
TSLA_data_kmc = []
TSLA_data_kmc = concatenated_data_kmc.drop(rare_short_words)
```

```python
TSLA_data_kmc = TSLA_data_kmc.transpose()
TSLA_data_kmc
```

## K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(TSLA_data_kmc)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

## Evaluation
### We compare these labels with the change_binary column in concatenated_data table 

```python
concatenated_data = concatenated_data.transpose()

from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(concatenated_data['Change_binary'],kmeans.labels_))
print(classification_report(concatenated_data['Change_binary'],kmeans.labels_))
```
