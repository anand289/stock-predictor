```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
concatenated_data = pd.read_excel("/Users/anandsingh/Desktop/Personal/Udemy_Courses/Web_Scraping/Tesla_Data/concatenated_table_tesla_1249.xlsx",index_col=0)
```

```python
concatenated_data.head()
```

```python
concatenated_data
```

```python
concatenated_data.dropna(axis=1,inplace=True)
concatenated_data
```

```python
# Removing words that don't appear for atleast 50 days or are smaller than two characters
minimum_occurance = 50 #minimum number of days a word should appear
minimum_word_length = 3
rare_short_words = []
for idx in concatenated_data.index[:-1]:
    if ((concatenated_data.loc[idx] == 0).sum()<minimum_occurance) or (len(idx)<minimum_word_length):
        rare_short_words.append(idx)
```

```python
TSLA_data = []
TSLA_data = concatenated_data.drop(rare_short_words)
```

```python
TSLA_data = TSLA_data.transpose()
TSLA_data
```

# Linear Regression

```python
TSLA_data_corr = TSLA_data.corrwith(TSLA_data['Change_stock_prices'])
```

```python
TSLA_data.columns[:-1]
```

```python
X = TSLA_data[TSLA_data.columns[:-1]]
```

```python
y = TSLA_data['Change_stock_prices']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
Coeff = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coeff'])
```

# Predictions

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
