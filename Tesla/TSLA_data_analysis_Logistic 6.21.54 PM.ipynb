```python
imported_data = pd.read_excel("/Users/anandsingh/Desktop/Personal/Udemy_Courses/Web_Scraping/Tesla_Data/concatenated_table_tesla_1249.xlsx",index_col=0)
```

```python
concatenated_data = imported_data
concatenated_data.head()
```

```python
concatenated_data
```

```python
concatenated_data.dropna(axis=1,inplace=True)
concatenated_data
```

```python
concatenated_data.drop(['Percentage_change'],axis=0,inplace =True)
```

```python
from nltk.corpus import stopwords
```

```python
'''
Drop:
1. Non-string
2. Non-alphanumeric
3. Stopwords
'''

drop_list = []
for idx in concatenated_data.index[:-2]:
    if type(idx) == float:
        drop_list.append(idx)
    if type(idx) == str:
        if idx.isalpha() == False:
            drop_list.append(idx)
        elif idx in stopwords.words('english'):
            drop_list.append(idx)
```

```python
bow = []
bow = concatenated_data.drop(drop_list)
```

```python
bow = bow.transpose()
bow
```

## Logistic Regression

```python
bow.columns[:-1]
```

```python
X = bow[bow.columns[:-1]]
```

```python
y = bow['Change_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

## Prediction

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
