```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
concatenated_data_svc = pd.read_excel("concatenated_table_1116_logistic.xlsx",index_col=0) 
#same data as used in logistic model will be used here
```

```python
concatenated_data_svc.head()
```

```python
concatenated_data_svc
```

```python
concatenated_data_svc.dropna(axis=1,inplace=True)
concatenated_data_svc
```

```python
concatenated_data_svc.drop(['Change_stock_prices'],axis=0,inplace =True)
```

```python
concatenated_data_svc
```

```python
# Removing words that don't appear for atleast 50 days or are smaller than two characters
minimum_occurance = 50 #minimum number of days a word should appear
minimum_word_length = 3
rare_short_words = []
for idx in concatenated_data_svc.index[:-1]:
    if ((concatenated_data_svc.loc[idx] == 0).sum()<minimum_occurance) or (len(idx)<minimum_word_length):
        rare_short_words.append(idx)
```

```python
TSLA_data_svm = []
TSLA_data_svm = concatenated_data_svc.drop(rare_short_words)
```

```python
TSLA_data_svm = TSLA_data_svm.transpose()
TSLA_data_svm
```

## Train Test Split

```python
X = TSLA_data_svm[TSLA_data_svm.columns[:-1]]
```

```python
y = TSLA_data_svm['Change_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Support Vector Machine with Default Values

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## SVM with Grid Search

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
# May take awhile!
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
