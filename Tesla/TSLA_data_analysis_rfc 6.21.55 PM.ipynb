```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
concatenated_data_rfc = pd.read_excel("concatenated_table_1116_logistic.xlsx",index_col=0) 
#same data as used in logistic model will be used here
```

```python
concatenated_data_rfc.head()
```

```python
concatenated_data_rfc
```

```python
concatenated_data_rfc.dropna(axis=1,inplace=True)
concatenated_data_rfc
```

```python
concatenated_data_rfc.drop(['Change_stock_prices'],axis=0,inplace =True)
```

```python
concatenated_data_rfc
```

```python
# Removing words that don't appear for atleast 50 days or are smaller than two characters
minimum_occurance = 50 #minimum number of days a word should appear
minimum_word_length = 3
rare_short_words = []
for idx in concatenated_data_rfc.index[:-1]:
    if ((concatenated_data_rfc.loc[idx] == 0).sum()<minimum_occurance) or (len(idx)<minimum_word_length):
        rare_short_words.append(idx)
```

```python
TSLA_data_rfc = []
TSLA_data_rfc = concatenated_data_rfc.drop(rare_short_words)
```

```python
TSLA_data_rfc = TSLA_data_rfc.transpose()
TSLA_data_rfc
```

## Random Forest Classifier

```python
X = TSLA_data_rfc[TSLA_data_rfc.columns[:-1]]
```

```python
y = TSLA_data_rfc['Change_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500)
```

```python
rfc.fit(X_train,y_train)
```

## Prediction 

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
